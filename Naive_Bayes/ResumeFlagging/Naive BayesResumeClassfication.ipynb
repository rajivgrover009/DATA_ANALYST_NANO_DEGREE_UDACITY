### Background:
In this exercise, we will use resume data and predict flagged vs Non-flagged values for a given resume. DataSet contains total of 125 Resume and we will use NaiveBayes algorithm to classify the Resumes.

## Import the Data Set

```python
import pandas as pd
df=pd.read_csv("DataScienceResumes.csv")
```

```python
del df['resume_id']
```

```python
df.head()
```

### Create Binary Labels

```python
df['label']=df['class'].map({'not_flagged':0,'flagged':1})
print(df.shape)
df.head()
```

### Convert Words in to Matrix format for frequency

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector=CountVectorizer()
```

### SPlit the data in to trianing and testing data

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['resume_text'],df['label'],
                                              random_state=1)
```

```python
count_vector
```

```python
print('Number of rows in the total set:{}'.format(df.shape[0]))
print('Number of rows in the training set:{}'.format(X_train.shape[0]))
print('Number of rows in the training set:{}'.format(X_test.shape[0]))
```

### Convert trainign and testing data in to Matrix format with word frequencies

```python
count_vector = CountVectorizer()
training_data=count_vector.fit_transform(X_train)
testing_data=count_vector.transform(X_test)
```

```python
print(count_vector.get_feature_names())
print(training_data.shape)
```

### Apply Naive Bayes Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data,y_train)
```

### Time to make prediction on Test Data

```python
predictions=naive_bayes.predict(testing_data)
```

### Evaluate Model

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print('Accuracy Score:',format(accuracy_score(y_test,predictions)))
print('Precision Score:',format(precision_score(y_test,predictions)))
print('Recall Score:',format(recall_score(y_test,predictions)))
print('F1 Score:', format(f1_score(y_test,predictions)))
```
